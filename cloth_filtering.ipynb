```python
# ROI 영역 잘못 잡힌 이미지 옮겨서 체크 (가로 길이 > 세로 길이)
from PIL import Image
import os
import shutil

def move_wrong_roi(folder_path, wrong_path):
    # 지정된 폴더에서 파일 목록을 가져오기
    files = os.listdir(folder_path)

    for file_name in files:
        file_path = os.path.join(folder_path, file_name)

        # 이미지 열기
        img = Image.open(file_path)        
        img.close()

        if img.width > img.height:
            new_path = os.path.join(wrong_path, file_name)
            shutil.move(file_path, new_path)
            print(f"Moved: {file_name} to {wrong_path}")

if __name__ == "__main__":
#     train_roi_folder = "E:/data_warehouse/Project/data/Training/roi_images/"
#     train_wrong_folder = "E:/data_warehouse/Project/data/Training/wrong_roi/"
#     move_wrong_roi(train_roi_folder, train_wrong_folder)
    
    test_roi_folder = "E:/data_warehouse/Project/data/Validation/roi_images/"
    test_wrong_folder = "E:/data_warehouse/Project/data/Validation/wrong_roi/"
    move_wrong_roi(test_roi_folder, test_wrong_folder)
```

```python
def move_images(src_folder, dest_folder):
    # A 폴더에서 파일 목록을 가져오기
    files_a = os.listdir(src_folder)
    
    # B 폴더에서 이미 있는 파일 목록 가져오기
    files_b = set(os.listdir(dest_folder))

    for file_name in files_a:
        file_path = os.path.join(src_folder, file_name)

        # B 폴더에 이미 같은 이름의 파일이 있는지 확인
        if file_name not in files_b:
            new_path = os.path.join(dest_folder, file_name)
            shutil.move(file_path, new_path)
            print(f"Moved: {file_name} to {dest_folder}")

if __name__ == "__main__":
    train_src_folder = "E:/data_warehouse/Project/data/Training/resize_images/"
    train_dest_folder = "E:/data_warehouse/Project/data/Training/roi_images/"
    
    move_images(train_src_folder, train_dest_folder)
    
    test_src_folder = "E:/data_warehouse/Project/data/Validation/resize_images/"
    test_dest_folder = "E:/data_warehouse/Project/data/Validation/roi_images/"
    
    move_images(test_src_folder, test_dest_folder)
```
